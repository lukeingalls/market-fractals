```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
class ticker:
    def __init__(self, filename, lim=None):
        self.df = pd.read_csv(filename, sep=',')
        self.index = 1 if self.df.columns.dtype != 'object' else self.df.columns[1]
        
        if (lim and lim < len(self.df)):
            self.df = self.df[-lim:-1]
        
    def plot_data(self):
        ticker_fig, ticker_ax = plt.subplots()
        
        ticker_ax.plot(self.df[self.index].values)
        
    
    def plot_variance(self):
        deltas = [0]
        for i in range(1, len(self.df)):
            deltas.append(self.df[self.index].values[i] - self.df[self.index].values[i-1])

        zscores = stats.zscore(deltas)
        
        delta_fig, delta_ax = plt.subplots()

        delta_ax.plot(zscores, 'b.', linestyle='None')
        delta_ax.set_ylim(-10, 10)
        
```

```python
SPY = ticker('SPY.csv')
SPY.plot_data()
SPY.plot_variance()
```

```python
GM = ticker('./GM.csv')
GM.plot_data()
GM.plot_variance()
```

```python
TSLA = ticker('./TSLA.csv', 1000)
TSLA.plot_data()
TSLA.plot_variance()
```
