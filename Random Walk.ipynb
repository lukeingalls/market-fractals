```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
ITERATIONS = 2188
STD = 1
MU = 0
```

```python
def pop_list_normal(MU, STD):
    global ITERATIONS
    ls = [0] * ITERATIONS
    ls[0] = np.random.normal(MU, STD)
    
    for i in range(1, ITERATIONS):
        ls[i] = ls[i - 1] + np.random.normal(MU, STD)
        
    return ls

def pop_list_cauchy():
    global ITERATIONS
    ls = [0] * ITERATIONS
    ls[0] = np.random.standard_cauchy()
    
    for i in range(1, ITERATIONS):
        ls[i] = ls[i - 1] + np.random.standard_cauchy()
        
    return ls
```

```python
normal_walk1 = pop_list_normal(MU, STD)
normal_walk2 = pop_list_normal(MU, STD)
normal_walk3 = pop_list_normal(MU, STD)
```

```python
x_vals = [i for i in range(ITERATIONS)]
```

```python
plt.plot(x_vals, normal_walk1)
plt.plot(x_vals, normal_walk2)
plt.plot(x_vals, normal_walk3)
plt.plot(x_vals, [MU for i in range(ITERATIONS)])
plt.title('Random walks with the normal distribution')
```

The graph above shows 3 random walks with MU and STD as the change for any particular moment.

```python
cauchy_walk1 = pop_list_cauchy()
cauchy_walk2 = pop_list_cauchy()
cauchy_walk3 = pop_list_cauchy()
```

```python
plt.plot(x_vals, cauchy_walk1)
plt.plot(x_vals, cauchy_walk2)
plt.plot(x_vals, cauchy_walk3)
plt.plot(x_vals, [MU for i in range(ITERATIONS)])
plt.title('Random walks with the cauchy distribution')
```

The graph above shows 3 random walks using the cauchy distribution

```python
deltas = [0]
for i in range(1, len(normal_walk1)):
    deltas.append(normal_walk1[i] - normal_walk1[i-1])
    
zscores = stats.zscore(deltas)
```

```python
delta_fig, delta_ax = plt.subplots()

delta_ax.plot(zscores, 'b.', linestyle='None')
delta_ax.set_ylim(-10, 10)
```
